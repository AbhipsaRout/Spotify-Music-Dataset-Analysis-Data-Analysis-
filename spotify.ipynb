```python
from sklearn.preprocessing import LabelEncoder
```

```python
data_copy = spotify.copy()
le = LabelEncoder()
```

```python
for col in data_copy.select_dtypes(include=['object']).columns:
    data_copy[col] = le.fit_transform(data_copy[col])
```

```python
x = data_copy.drop('popularity', axis=1)   # input
y = data_copy['popularity'] 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)   
x_test = sc.transform(x_test) 
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
ypred = lr.predict(x_test)
ypred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
r2=r2_score(y_test,ypred)
mse=mean_squared_error(y_test,ypred)
print('r2_score is:',r2)
print('mean squared error is:',mse)
```

```python
df=pd.DataFrame({'prediction':ypred,'actual data':y_test})
sns.lmplot(df,x='prediction',y='actual data')
plt.show()
```
